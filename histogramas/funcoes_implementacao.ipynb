```python
from skimage.io import imread, imshow, imsave
from skimage.color import rgb2gray
import numpy as np
import cv2
```

```python
image = imread('../training/glioma/glioma_29.jpg')
image = rgb2gray(image)
```

```python
def negativo(image):
  return 255 - image
```

```python
image_ne = negativo(image)
```

```python
imshow(image_ne, cmap='gray')
```

```python
def potencia(image, gamma):
  return (image / 255) ** gamma * 255
```

```python
def power_law(image, gamma):
  return 255 * np.power(image / 255, gamma)
```

```python
def handfun(image, gamma):
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            image[i][j] = (image[i][j] / 255) ** gamma * 255
    return image
```

```python
image_f1 = potencia(image, 10)
image_f2 = power_law(image, 0.5)
image_f3 = handfun(image, 0.5)
```

```python
imshow(image_f1, cmap='gray')
```

```python
imshow(image_f2, cmap='gray')
```

```python
imshow(image_f3, cmap='gray')
```

A transformao logartmica  uma transformao matemtica que pode ser usada em imagens ou sinais para compresso de dados, reduo de rudo e outros fins. A funo dada :

s = c * log(1 + r)

onde s  a sada da transformao, r  a entrada (a imagem ou sinal) e c  um parmetro de escala.

O parmetro c  um escalar que controla a intensidade da transformao logartmica. Ele pode ser escolhido de acordo com as necessidades especficas de cada aplicao.

Em geral, o valor de c  escolhido de modo a equilibrar a compresso de dinmica (a reduo do intervalo de valores da imagem) com a preservao de detalhes (a manuteno de pequenas variaes de brilho na imagem). Valores maiores de c resultam em maior compresso de dinmica, enquanto valores menores resultam em menor compresso de dinmica e maior preservao de detalhes.

Para escolher o valor de c,  comum experimentar com diferentes valores e avaliar o resultado da transformao em relao s necessidades da aplicao. Alguns valores tpicos para c so: c = 1, c = 10 e c = 100.

```python
# c = 255/(np.log(1 + np.max(image)))
def logaritmica(image, c):
  return c * np.log(1 + image)
```

```python
#image_log1 = logaritmica(image, 0.5).astype('uint8')
image_log1 = logaritmica(image, 0.5)
```

```python
imshow(image_log1, cmap='gray')
```

```python
def Alargamento_de_contraste(image, k, E):
  return 1/(1 + (k/image)**E)
```

```python
image_ac = Alargamento_de_contraste(image, 150, 3)
```

```python
imshow(image_ac, cmap='gray')
```
